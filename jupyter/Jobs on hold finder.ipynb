```python
import sys
sys.path.insert(0, "../app/modules")
from servicetitan_api_client import ServiceTitanClient
from data import get_secret, state_codes
import pandas as pd
```

```python
state = 'NSW'
state_code = state_codes()[state]

client = ServiceTitanClient(app_key=get_secret("ST_app_key_tester"), tenant=get_secret(f"ST_tenant_id_{state_code}"), client_id=get_secret(f"ST_client_id_{state_code}"), client_secret=get_secret(f"ST_client_secret_{state_code}"), environment="production")
```

```python
url = client.build_url('jpm', 'jobs')
url
```

```python
hold_jobs = client.get_all(url, params={'jobStatus': 'Hold'})
```

```python
hold_jobs_fmt = []
for job in hold_jobs:
    fmt = {
        'Job Number': job['jobNumber'],
        'Status': job['jobStatus'],
        'Date Created': client.format_local(client.from_utc(job['createdOn']), fmt="%Y-%m-%d %H:%M"),
    }
    hold_jobs_fmt.append(fmt)
```

```python
df = pd.DataFrame(hold_jobs_fmt)
df.head()
```

```python
df.to_csv(f'jobs_on_hold_{state_code}_as_of_20251107.csv', index=False)
```

```python
big_df = pd.DataFrame()
for state in ['NSW', 'NSW_old', 'VIC_old', 'QLD_old', 'WA']:
    state_code = state_codes()[state]

    client = ServiceTitanClient(app_key=get_secret("ST_app_key_tester"), tenant=get_secret(f"ST_tenant_id_{state_code}"), client_id=get_secret(f"ST_client_id_{state_code}"), client_secret=get_secret(f"ST_client_secret_{state_code}"), environment="production")
    url = client.build_url('jpm', 'jobs')
    hold_jobs = client.get_all(url, params={'jobStatus': 'Hold'})

    hold_jobs_fmt = []
    for job in hold_jobs:
        fmt = {
            'Job Number': job['jobNumber'],
            'Status': job['jobStatus'],
            'Date Created': client.format_local(client.from_utc(job['createdOn']), fmt="%Y-%m-%d"),
            'Time Created': client.format_local(client.from_utc(job['createdOn']), fmt="%H:%M"),
            'Tenant Codename': state_code
        }
        hold_jobs_fmt.append(fmt)
    df = pd.DataFrame(hold_jobs_fmt)
    big_df = pd.concat([big_df, df])
    df.to_csv(f'jobs_on_hold_{state_code}_as_of_20251107.csv', index=False)
```

```python
big_df.to_csv(f'jobs_on_hold_total_as_of_20251107.csv', index=False)
```
