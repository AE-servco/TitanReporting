```python
# Combine start and end times into tuples
timesheet_data['times'] = list(zip(timesheet_data['arrivedTime'], timesheet_data['doneTime']))

# Ensure date column is datetime (for proper sorting)
# timesheet_data['date'] = pd.to_datetime(timesheet_data['arrivedDate'])

# Pivot so names become columns, dates become rows, and values are tuples
reshaped = (
    timesheet_data
    .pivot(index='arrivedDate', columns='tech_name', values='times')
    .sort_index(axis=0)    # sort by date
    .sort_index(axis=1)    # sort columns alphabetically by name
    .reset_index()
)
```
